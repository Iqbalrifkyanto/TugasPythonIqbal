```python
# Import library yang dibutuhkan
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
url = "MBP.csv"
df = pd.read_csv(url)

df.drop(columns=["Back/Rare Camera","Front Camera","Battery"],inplace=True)

df.head()
```

```python
# Informasi dataset
df.info()

# Statistik deskriptif
df.describe()

# Cek missing values
print(df.isnull().sum())
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
numerial_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', MinMaxScaler())
])

Categorical_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore'))
])

preprocessing = ColumnTransformer([
    ('rating', numerial_pipeline),
])
a = Pipeline([
    ('preprocessing', preprocessing),
    ('model', KNeighborsClassifier())
])
y = df['Price in INR']
numbers = [int(item.replace('₹', '').replace(',', '')) for item in y]
x = [int(item.replace('₹', '').replace(',', '')) for item in df['Number of Ratings']]
x_train, x_test, y_train, y_test = train_test_split(x, numbers, test_size=0.2, random_state=42)
if 'Price in INR' in df.columns:
    print("Kolom 'Price in INR' ditemukan.")
else:
    print("Kolom 'Price in INR' tidak ditemukan.")
# Mengubah x_train dan x_test menjadi array 2D
x_train = np.array(x_train).reshape(-1, 1)
x_test = np.array(x_test).reshape(-1, 1)

# Melatih model dengan data yang sudah diubah menjadi 2D
kmn = KNeighborsClassifier(n_neighbors=3)
kmn.fit(x_train, y_train)
kmn.score(x_test, y_test)
```
