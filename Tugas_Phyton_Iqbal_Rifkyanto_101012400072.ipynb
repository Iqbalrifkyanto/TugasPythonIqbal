```python
# Import library yang dibutuhkan
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "Employee.csv"
df = pd.read_csv(url)
df.head()

```

```python
# Informasi dataset
df.info()

# Statistik deskriptif
df.describe()

# Cek missing values
print(df.isnull().sum())
```

```python
# Scatter plot hubungan pengalaman kerja vs. umur
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['ExperienceInCurrentDomain'], y=df['Age'])
plt.xlabel('Experience In Current Domain')
plt.ylabel('Age')
plt.title('Pengalaman Kerja vs Umur')
plt.show()

```

Scatter Plot menunjukan semakin tua umur semakin banyak pengalaman kerjanya
